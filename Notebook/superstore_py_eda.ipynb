### Feature Engineering

```python
#House age, the overall time since when the house has been built
from datetime import datetime
current_year = datetime.now().year

df['House_age'] = current_year - df['Yr_built']
```

```python
df[['House_age', 'Id', 'Yr_built']]  
```

```python
df['Has_Basement'] = df['Sqft_basement'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df[['Id', 'Has_Basement', 'Sqft_basement']]  
```

```python
#Years since rennovation highlights the time period it has been since the house was renovated, if it hasnt been renovated it would return the time period the house has been around 

df['YearsSinceRenovation'] = np.where(df['Yr_renovated'] > 0, current_year - df['Yr_renovated'],current_year - df['Yr_built'])
df['Has_Renovation'] = df['Yr_renovated'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df[["Yr_built","Yr_renovated","YearsSinceRenovation"]]
```

```python
#price per square foot is an imp aspect when determining or evaluating properties
df['Price_persqft'] = df['Price'] / df['Sqft_living']
```

```python
df[['Id', 'Price','Sqft_living', 'Price_persqft']]  
```

```python
# Compare house size with neighborhood average, =1 is well equal to the neighbourhood avg,<1 is smaller than area avg so it can be priced lower,>1 it can be priced higher cuz of its area avg compatibility
df['Relative_size'] = df['Sqft_living'] / df['Sqft_living15']
```

```python
df[['Id', 'Price','Sqft_living', 'Relative_size']]  
```

```python
sns.scatterplot(x='Relative_size', y='Price', data=df)
```

```python
df_filtered = df[df['Relative_size'] < 4]
```

```python
sns.scatterplot(x='Relative_size', y='Price', data=df_filtered)
```

```python
df['Relativesizecategory'] = pd.cut(df['Relative_size'],bins=[0,0.8,1.5,4],labels=['Small','Average','Large'])
```

```python
df.groupby('Relativesizecategory')['Price'].mean()
```
